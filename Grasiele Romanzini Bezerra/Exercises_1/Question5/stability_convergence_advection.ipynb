In this notebook I explore numerically the stability condition for the advection equation

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

The stability condition for the advection equation is given by
$$
CFL =  \nu a \leq   \frac{1}{(1 - 2\Theta)}.
$$

```python
def CFL_max_condition(theta):
    return 1/(1-2.0*theta)
```

```python
theta_values = np.linspace(0, 0.49, num=50)
CFL_number = []
for theta in theta_values:
    CFL_number.append(CFL_max_condition(theta))
```

```python
sns.set_context("paper")

h_plot = 1
v_plot = 1

fig, ax = plt.subplots(v_plot, h_plot,  figsize=(8, 6))

sns.lineplot(x=theta_values, y=CFL_number, color = "orange")
ax.axvline(x = 0.5, ls = 'dashed', c = 'purple') 
# ax.axhline(y = 0,  c = 'gray')
ax.set_xlabel(r"$\Theta$")
ax.set_ylabel(r"CFL Numeber")
sns.despine(top=True, right=True, left=False, bottom=False)
ax.set_xlim([0, 0.8])
ax.set_ylim([0, 10])
ax.set_xticks([0.5])
ax.set_yticks([0,1])
# ax.set_xticklabels([r"1/2"])
ax.text(0.2,4, 'unstable', rotation=45,fontsize=15, color = 'red')
ax.text(0.45,4, 'stable', rotation=45,fontsize=15, color = 'blue')
plt.savefig('images/Question5-stability.png', dpi = 500, bbox_inches = 'tight') #facecolor='white'
```
